## 사용 라이브러리

```python
import pandas as pd
import numpy as np
import copy
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.family'] = 'Malgun Gothic'
```

## 사용데이터

```python
df = pd.read_csv('C:\\workspace\\Next_Ez_Training\\data\\golflike\\골프장데이터_해당년도_추가.csv')
df.head(3)
```

## 해비치와 라온

```python
plt.subplot(211)# 2 x 2, 그래프중 첫번째
plt.ylim(0,500)
plt.bar(df[df.사용일자==202005.0].groupby(df.골프장명).골프장명.count().keys(), df[df.사용일자==202005.0].groupby(df.골프장명).골프장명.count().values)
plt.subplot(2, 1, 2)
plt.bar(df[df.사용일자==202105.0].groupby(df.골프장명).골프장명.count().keys(), df[df.사용일자==202105.0].groupby(df.골프장명).골프장명.count().values)
plt.show()
```

- 해비치CC 와 라온 GC 경우에 큰 변화가 있는것을 확인하였다

## 년도별로 분석

```python
plt.figure(figsize=(30,10))
plt.subplot(211)# 2 x 2, 그래프중 첫번째
plt.ylim(0,3000)
plt.bar(df[df.해당년도==2020.0].groupby(df.골프장명).골프장명.count().keys(), df[df.해당년도==2020.0].groupby(df.골프장명).골프장명.count().values)
plt.subplot(2, 1, 2)
plt.bar(df[df.해당년도==2021.0].groupby(df.골프장명).골프장명.count().keys(), df[df.해당년도==2021.0].groupby(df.골프장명).골프장명.count().values)
plt.show()
```

## 월별로 분석

```python
plt.subplot(211)# 2 x 2, 그래프중 첫번째
plt.ylim(0,500)
plt.bar(df[df.사용일자==202005.0].groupby(df.골프장명).골프장명.count().keys(), df[df.사용일자==202005.0].groupby(df.골프장명).골프장명.count().values)
plt.subplot(2, 1, 2)
plt.bar(df[df.사용일자==202007.0].groupby(df.골프장명).골프장명.count().keys(), df[df.사용일자==202007.0].groupby(df.골프장명).골프장명.count().values)
plt.show()
```
