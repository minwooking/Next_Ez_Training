```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
mpl.rcParams['axes.unicode_minus'] = False
[(f.name, f.fname) for f in fm.fontManager.ttflist if 'Malgun' in f.name]
plt.rcParams['font.family'] = 'Malgun Gothic'
import seaborn as sns
```

## 제주 인구 유입

```python
years = '2010년	2011년	2012년	2013년	2014년	2015년	2016년	2017년	2018년	2019년	2020년	2021년 \t2022년'.split('\t')
humans = '577,187	583,284	592,449	604,670	621,550	641,355	661,190	678,772	692,032\t696,657	697,578	697,476'.split('\t')
data=dict(zip(years,humans))
s = pd.Series(data)
s = pd.DataFrame(s)
s=s.apply(lambda x: x.str.replace(',',''))
s= s.T
s = s.apply(lambda x : int(x) if np.dtype(x) == object else x)
s = pd.DataFrame(s)
s.rename(columns={0 : '유입인구수'},inplace=True)
s.plot()
plt.grid()
```

## 고령인구비율

```python
old_years = '2010년	2011년	2012년	2013년	2014년	2015년	2016년	2017년	2018년	2019년	2020년	2021년'.split('\t')
old_humans = '12.1	12.5	12.8	13.1	13.3	13.4	13.5	13.8	13.9	14.5	15.3	15.9'.split('\t')
```

```python
data=dict(zip(years,old_humans))
old_S = pd.Series(data)
old_S = pd.DataFrame(old_S)
old_S=old_S.apply(lambda x: x.str.replace(',',''))
old_S= old_S.T
old_S = old_S.apply(lambda x : float(x) if np.dtype(x) == object else x)
old_S = pd.DataFrame(old_S)
old_S.rename(columns={0 : '고령자증가율'},inplace=True)
old_S.plot()
plt.grid()
```

## 1인가구증가율

```python
onehome_years = '2015년	2016년	2017년	2018년	2019년	2020년'.split('\t')
onehome_humans = '58,446	62,841	68,738	73,250	75,353	81,855'.split('\t')
data=dict(zip(onehome_years,onehome_humans))
onehome = pd.Series(data)
onehome = pd.DataFrame(onehome)
onehome=onehome.apply(lambda x: x.str.replace(',',''))
onehome= onehome.T
onehome = onehome.apply(lambda x : int(x) if np.dtype(x) == object else x)
onehome = pd.DataFrame(onehome)
onehome.rename(columns={0 : '1인가구수'},inplace=True)
onehome.plot()
plt.grid()
```

## 60대 골프장 선호도 조사

```python
df.groupby(df.골프장명).골프장명.count().plot.barh(figsize=(24,10),
                                                                color='hotpink',use_index=True,
                                                                 )
plt.xticks(label=df.골프장명)
plt.grid()
# plt.color
plt.title('60대 골프장 선호 조사')
plt.show()
# plt.
# plt.xscale
# plt.savefig('test.png', dpi = 100)
```
